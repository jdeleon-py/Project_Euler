<p>The smallest positive integer $n$ for which the numbers $n^2 + 1$, $n^2 + 3$, $n^2 + 7$, $n^2 + 9$, $n^2 + 13$, and $n^2 + 27$ are consecutive primes is $10$. The sum of all such integers $n$ below one-million is $1242490$.</p>

<p>What is the sum of all such integers $n$ below $150$ million?</p>

```python
def is_prime(n: int) -> bool:
    """Return True if n is prime, otherwise False."""
    if n < 2:
        return False

    if n == 2:
        return True

    if n % 2 == 0:
        return False

    # Only test odd divisors up to sqrt(n)
    divisor = 3
    while divisor * divisor <= n:
        if n % divisor == 0:
            return False
        divisor += 2
    return True

if __name__ == "__main__":
    print(is_prime(19))
    print(is_prime(20))
```

```python
cond1: bool = lambda n: True if is_prime((n ** 2) + 1) else False
cond3: bool = lambda n: True if is_prime((n ** 2) + 3) else False
cond7: bool = lambda n: True if is_prime((n ** 2) + 7) else False
cond9: bool = lambda n: True if is_prime((n ** 2) + 9) else False
cond13: bool = lambda n: True if is_prime((n ** 2) + 13) else False
cond27: bool = lambda n: True if is_prime((n ** 2) + 27) else False

num_sum: int = 0
for i in range(3, 1000001):
    if cond1(i) and cond3(i) and cond7(i) and cond9(i) and cond13(i) and cond27(i):
        print(i)
        num_sum += i

# use optimized C program for faster results

# NOTE: n = 2 also works??
```

```python
# Python3 program to find the smallest
# twin in given range 
import random

# Iterative Function to calculate 
# (a^n) % p in O(logy) 
def power(a: int, n: int, p: int): 
    res: int = 1                     # Initialize result
    a = a % p                        # Update 'a' if 'a' >= p 
    while n > 0:
        if n % 2 == 1:               # if n is odd, multiply result by 'a'
            res = (res * a) % p
            n -= 1
        else:                        # n must be even now
            a = (a ** 2) % p 
            n = n // 2
    return res % p
    
# If n is prime, then always returns true,
# If n is composite than returns false with
# high probability Higher value of k increases
# probability of correct result
def isPrime(n: int, k: int):
    if n == 1 or n == 4:
        return False
    elif n == 2 or n == 3:
        return True
    else:
        for i in range(0, k): # Try k times 
            # Pick a random number 
            # in [2..n-2]      
            a = random.randint(2, n - 2)
            
            # Fermat's little theorem 
            if power(a, n - 1, n) != 1:
                return False            
    return True
            
# Driver code
k = 3
print(True) if isPrime(11, k) else print(False)
print(True) if isPrime(21, k) else print(False)
```
